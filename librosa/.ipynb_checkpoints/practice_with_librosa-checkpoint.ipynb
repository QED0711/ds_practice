```python
def get_audio_chroma(y, sr, start_stop=[0, 60], x_bins=7*3*12, y_bins=12*3):
    chroma_orig = librosa.feature.chroma_cqt(y=y, sr=sr)
    idx = tuple([slice(None), slice(*list(librosa.time_to_frames(start_stop)))])

    C = np.abs(librosa.cqt(y=y, sr=sr, bins_per_octave=x_bins, n_bins=y_bins))

    return pd.DataFrame(C)
    
```

# Spectrogram

```python
sample_rate = 22050
fft_size = 256
```

```python
# Using y_orig imported above
orig_spec_df = pd.DataFrame(librosa.feature.melspectrogram(y=y_orig, sr=sample_rate, n_mels=fft_size))
mod_spec_df = pd.DataFrame(librosa.feature.melspectrogram(y=y_mod, sr=sample_rate, n_mels=fft_size))
```

```python
orig_spec_df.mean(axis=1).loc[:50].plot()
mod_spec_df.mean(axis=1).loc[:50].plot()
```

```python
def create_averaged_df(df):
    # get the mean of the column values
    averaged = df.mean(axis=1)
    # scale the column values between 0. and 1.
#     averaged = np.interp(averaged, (averaged.min(), averaged.max()), (0., 1.))
    
    averaged_df = pd.DataFrame(averaged, columns=['average_amplitude'])
#     print(averaged_df)
    return averaged_df

def add_freq_bins(df, sample_rate, fft_size):
    df['freq_bins'] = np.NaN
    for i in range(len(df.freq_bins)):
        df.freq_bins[i] = i * sample_rate / fft_size
        
    return df
```

```python
def find_ratio(x):
    return x['average_amplitude'] / x['orig_amp']

orig_df = create_averaged_df(orig_spec_df)
orig_with_freq = add_freq_bins(orig_df, sample_rate, fft_size)

mod_df = create_averaged_df(mod_spec_df)
mod_with_freq = add_freq_bins(mod_df, sample_rate, fft_size)

mod_with_freq['orig_amp'] = orig_with_freq['average_amplitude']
mod_with_freq['ratios'] = mod_with_freq.apply(find_ratio, axis=1)

mod_with_freq.tail(10)
```

```python

plt.figure(figsize=[12, 4])

bins = 75

orig_with_freq.loc[:bins].plot(kind="line", x="freq_bins", y='average_amplitude')
plt.xlabel('Frequency (Hz)')
plt.ylabel("Aplitude")
plt.title('Original')
plt.ylim(0, 150)

mod_with_freq.loc[:bins].plot(kind="line", x="freq_bins", y='average_amplitude', color="red")
plt.xlabel('Frequency (Hz)')
plt.ylabel("Aplitude")
plt.title('Modified')
plt.ylim(0, 150)

mod_with_freq.loc[:bins].plot(kind="line", x="freq_bins", y='ratios', color="green")
plt.xlabel('Frequency (Hz)')
plt.ylabel("Aplitude")
plt.title('Ratio to Ideal')
plt.ylim(0, 3)

```

```python
def spec_display(S, idx, max_freq=8000):
    return librosa.display.specshow(librosa.power_to_db(S, ref=np.max)[idx], 
                                    y_axis='mel', fmax=10_000, x_axis='time')
```

```python

S_orig = librosa.feature.melspectrogram(y=y_orig, n_mels=512, fmax=10_000)
S_mod = librosa.feature.melspectrogram(y=y_mod, n_mels=512, fmax=10_000)

idx = tuple([slice(None), slice(*list(librosa.time_to_frames([20, 25])))])

plt.figure(figsize=(10, 4))

plt.subplot(2, 1, 1)
# librosa.display.specshow(librosa.power_to_db(S_orig, ref=np.max)[idx], y_axis='mel', fmax=10_000, x_axis='time')
spec_display(S_orig, idx)
plt.colorbar(format='%+2.0f dB')
plt.title("Original Audio")
plt.xlabel("Time (seconds)")

plt.subplot(2,1,2)
# librosa.display.specshow(librosa.power_to_db(S_mod, ref=np.max)[idx], y_axis='mel', fmax=10_000, x_axis='time')
spec_display(S_mod, idx)
plt.colorbar(format='%+2.0f dB')
plt.title("Modified Audio")
plt.xlabel("Time (seconds)")

plt.tight_layout()
```
