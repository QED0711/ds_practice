```python
import librosa
from librosa.feature import chroma_stft
import librosa.display

import soundfile as sf

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# filename = librosa.util.example_audio_file()

# import as .wav file
filename_original = 'ds_test_original.wav'
filename_modified = 'ds_test_bass_modified.wav'
# load in audio files, 
# y = waveform
# sr = sample rate

y_orig, sr_orig = librosa.load(filename_original)

y_mod, sr_mod = librosa.load(filename_modified)
```

```python
def get_audio_chroma(y, sr, start_stop=[0, 60], x_bins=7*3*12, y_bins=12*3):
    chroma_orig = librosa.feature.chroma_cqt(y=y, sr=sr)
    idx = tuple([slice(None), slice(*list(librosa.time_to_frames(start_stop)))])

    C = np.abs(librosa.cqt(y=y, sr=sr, bins_per_octave=x_bins, n_bins=y_bins))

    return pd.DataFrame(C)
    
```

```python
chroma_orig = librosa.feature.chroma_cqt(y=y, sr=sr)

# For display purposes, let's zoom in on a 15-second chunk from the middle of the song
idx = tuple([slice(None), slice(*list(librosa.time_to_frames([45, 60])))])

# And for comparison, we'll show the CQT matrix as well.
C = np.abs(librosa.cqt(y=y, sr=sr, bins_per_octave=12*3, n_bins=7*3*12))

chroma_df = pd.DataFrame(C)

plt.figure(figsize=(12, 4))
plt.subplot(2, 1, 1)
librosa.display.specshow(librosa.amplitude_to_db(C, ref=np.max)[idx],
                         y_axis='cqt_note', bins_per_octave=12*3)
plt.colorbar()
plt.subplot(2, 1, 2)
librosa.display.specshow(chroma_orig[idx], y_axis='chroma')
plt.colorbar()
plt.ylabel('Original')
plt.tight_layout()
```
