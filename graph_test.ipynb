```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.array([x for x in range(10)])

a = x
b = x**2
c = x**3
d = x**4

a_df = pd.DataFrame(a)
b_df = pd.DataFrame(b)
c_df = pd.DataFrame(c)
d_df = pd.DataFrame(d)
```

```python
plt.plot(a_df)
plt.plot(b_df)
plt.plot(c_df)
plt.plot(d_df)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)

ax1.plot(a)
ax1.set_title("A")

ax2.plot(b)
ax2.set_title("B")

ax3.plot(c)
ax3.set_title("C")

ax4.plot(d)
ax4.set_title("D")

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(10,10))

plt.subplot2grid((3, 3), (0, 0), rowspan=2, colspan=2)
plt.subplot2grid((3, 3), (2, 0), rowspan=1, colspan=3)
plt.subplot2grid((3, 6), (0, 4), rowspan=2, colspan=1)
plt.subplot2grid((5, 6), (0, 5), rowspan=2, colspan=1)
```

```python

def annotate_axes(fig):
    for i, ax in enumerate(fig.axes):
        ax.text(0.5, 0.5, "ax%d" % (i+1), va="center", ha="center")
        ax.tick_params(labelbottom=False, labelleft=False)


fig = plt.figure()

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=3)
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)
ax4 = plt.subplot2grid((3, 3), (2, 0))
ax5 = plt.subplot2grid((3, 3), (2, 1))
ax0 = plt.subplot2grid((2,2), (1,1))
annotate_axes(fig)

plt.show()
```
