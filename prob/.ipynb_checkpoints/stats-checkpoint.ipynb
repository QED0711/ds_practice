## From different populations

```python
samp1 = [np.random.normal(loc=0, scale=1) for x in range(100)]

samp2 = [np.random.normal(loc=0, scale=2) for x in range(100)]

sns.distplot(samp1, hist=False)
sns.distplot(samp2, hist=False)

stats.ttest_ind(samp1, samp2)
```

```python

def check_h(alpha, mean, std):
    df = pd.DataFrame(columns=["reject_null", "p_value", "alpha"])

    for i in range(100):

        samp1 = [np.random.normal(loc=0, scale=1) for x in range(100)]
        samp2 = [np.random.normal(loc=mean, scale=std) for x in range(100)]
        (ts, pv) = stats.ttest_ind(samp1, samp2)

    #         print(pv < alpha)
        if pv < alpha:
            df = df.append({'reject_null': False, "p_value": pv, "alpha": alpha}, ignore_index=True)
        else:
            df = df.append({'reject_null': True, "p_value": pv, "alpha": alpha}, ignore_index=True)
    return df
        
```

```python
df = check_h(0.05, 1, 1)

plt.bar(['reject', 'fail_reject'], [df[df.reject_null].shape[0], df[df.reject_null == False].shape[0]])
print([df[df.reject_null].shape[0], df[df.reject_null == False].shape[0]])
```
